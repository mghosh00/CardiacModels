# Inferring parameters for the FitzHugh-Nagumo Model

In this notebook, we use PINTS to create noisy data for a FitzHugh-Nagumo system and then perform inference on the model
to try and retrieve the parameters.

## Creating the ForwardModel

Firstly, we set up the system of ODEs for the FitzHugh-Nagumo system. Below are the non-dimensional equations we will use for a fast variable, $v$ - the membrane potential, and a slow variable, $w$ - the ion current. Note that these equations are taken directly from Keener and Sneyd (1998) Section 4.2 [^1].

$v$ - the membrane potential, fast variable

$w$ - the ion current, slow variable

\begin{equation}
    \epsilon \frac{dv}{dt} = Av(v - \alpha)(1 - v) - w,
\end{equation}
\begin{equation}
    \frac{dw}{dt} = v - \gamma w,
\end{equation}
\begin{equation}
    v(0) = v_{0}, \qquad w(0) = w_{0}.
\end{equation}

The goal is to try and initially infer $\epsilon$, $A$, $\alpha$, $\gamma$, $v_{0}$ and $w_{0}$, and then
see which parameters we can restrict to improve the overall inference.

```python
import pints
import numpy as np
import scipy.integrate as si
import matplotlib.pyplot as plt
```

### PINTS implementation of the model

In addition to implementing the system, we give the option in the class to specify fixed parameters, if we wish to 
reduce the dimensions of the parameter space.

```python
class FitzHughNagumoModel(pints.ForwardModel):

    def __init__(self, fixed_params=None, num_outputs=2):
        """Initialises the forward model with certain fixed parameters passed to the method
        as key-value pairs.
        """
        assert num_outputs in [1, 2]
        self._num_outputs = num_outputs
        super(FitzHughNagumoModel, self).__init__()
        self._param_names = ["epsilon", "A", "alpha", "gamma", "v0", "w0"]
        self._fixed_params = {}
        if fixed_params is not None:
            for k, v in fixed_params:
                if k in self._param_names:
                    self._fixed_params[k] = v

    def get_param_names(self):
        """Returns the parameter names.
        """
        return self._param_names

    def n_outputs(self):
        """Returns number of model outputs, i.e. v and w for 2 and just v for 1
        """
        return self._num_outputs

    def n_parameters(self):
        """Returns number of parameters to be inferred, i.e. all non-fixed parameters.
        """
        return 6 - len(self._fixed_params)

    def n_fixed_parameters(self):
        """Returns number of parameters we fix.
        """
        return len(self._fixed_params)

    def simulate(self, parameters, times):
        """This inherits from pints.ForwardModel, performs one forward simulation of the model.
        The list `parameters` contains only the parameters we are inferring.
        """
        epsilon, A, alpha, gamma, v0, w0 = self._get_all_params(parameters)
        t_span = [times[0], times[len(times) - 1]]
        y0 = [v0, w0]
        sol = si.solve_ivp(fun=self._rhs, t_span=t_span, y0=y0, 
                           t_eval=times, args=(epsilon, A, alpha, gamma))

        # If we have 1 output, then we just want to return v. Otherwise, return both v and w
        return sol.y.transpose()[:, :self._num_outputs]

    def _get_all_params(self, parameters):
        """We need a list of 6 parameters, so we check to see which parameters are fixed, and which are not.
        We increment j each time a non-fixed parameter is reached.
        """
        all_parameters = []
        j = 0
        for param in self._param_names:
            if param in self._fixed_params.keys():
                all_parameters.append(self._fixed_params[param])
            else:
                all_parameters.append(parameters[j])
                j += 1
        return tuple(all_parameters)

    @staticmethod
    def _rhs(t, y, epsilon, A, alpha, gamma):
        """The right hand side function of the FitzHugh-Nagumo system.
        """
        v, w = y[0], y[1]
        return np.array([(A * v * (v - alpha) * (1 - v) - w) / epsilon,
                         v - gamma * w])

    def plot_nullclines(self, parameters, xlim, ylim):
        epsilon, A, alpha, gamma, v0, w0 = self._get_all_params(parameters)
        t = 0.0
        v = np.linspace(-1, 1, 100)
        ax = plt.gca()
        ax.set_xlim(xlim)
        ax.set_ylim(ylim)
        plt.plot(v, v / gamma, '--r', label='w nullcline', alpha=0.7)
        plt.plot(v, A * v * (v - alpha) * (1 - v), '--g', label='v nullcline', alpha=0.7)
        plt.xlabel('v')
        plt.ylabel('w')
```

## Generating data for the model

### Simulating the system

Next, we will generate some data for the model using known parameter values. We use the ones chosen by Keener and Sneyd (pg 140) [^1], tabulated below:

|parameter|value|
|-----|-----|
|$\epsilon$|0.01|
|$A$|1.0|
|$\alpha$|0.1|
|$\alpha_{osc}$|-0.1|
|$\gamma$|0.5|
|$v(0)$|0.2|
|$w(0)$|0.0|

We will use $\alpha_{osc}$ if we are trying to infer oscillations, otherwise we will just use $\alpha$.

```python
fixed_params = None
```

```python
model = FitzHughNagumoModel(fixed_params=fixed_params, num_outputs=1)
param_names = model.get_param_names()
parameters = np.array([0.01, 1.0, -0.1, 0.5, 0.2, 0.0])
times = np.arange(0, 3, 0.03)
data = model.simulate(parameters, times)
```

### Adding noise to the system

We now utilise PINTS noise generators to create some noisy data for the system. Specifically, `pints.noise.independent`
creates independent Gaussian noise iid $\mathcal{N}(0, \sigma)$.

```python
sigma = 0.01
noisy_data = data + pints.noise.independent(sigma, data.shape)
```

## Using PINTS to perform inference

We will now perform Bayesian inference on the model to try and recover the values of the input parameters.

### `LogPrior`, `LogLikelihood` and `LogPosterior`

```python
problem = pints.MultiOutputProblem(model, times, noisy_data)
```

```python
log_prior = pints.ComposedLogPrior(
    pints.UniformLogPrior([0.001], [0.1]), # epsilon
    pints.TruncatedGaussianLogPrior(1, 1, 0, np.inf), # A
    pints.UniformLogPrior([-1], [0]), # alpha (between -1 and 0)
    pints.TruncatedGaussianLogPrior(1, 1, 0, np.inf), # gamma
    pints.TruncatedGaussianLogPrior(0, 1, 0, np.inf), # v0
    pints.TruncatedGaussianLogPrior(0, 1, 0, np.inf), # w0
    pints.TruncatedGaussianLogPrior(0, 1, 0, np.inf), # sigma_v
    # pints.TruncatedGaussianLogPrior(0, 1, 0, np.inf) # sigma_w
)
```

```python
log_likelihood = pints.GaussianLogLikelihood(problem)
log_posterior = pints.LogPosterior(log_likelihood, log_prior)
```

### Initialising chains

```python
num_chains = 4
xs = log_prior.sample(num_chains)
```

### Performing inference

```python
n_params = log_prior.n_parameters()
transform = pints.RectangularBoundariesTransformation([0.001, 0, -4, 0, 0, 0, 0], 
                                                      [4, 4, 0, 4, 4, 4, 4])
mcmc = pints.MCMCController(log_posterior, num_chains, xs, method=pints.PopulationMCMC, transformation=transform)
max_iterations = 80000
mcmc.set_max_iterations(max_iterations)

chains = mcmc.run()
```

## Extracting posterior distributions

Finally, we will use the stationary chain data to extract posteriors. Assuming chains have converged, we can use their 
iterations as samples from the posterior.

```python
chains_df = pd.DataFrame(columns=all_param_names + ['chain'])
for i in range(num_chains):
    chain_i_df = pd.DataFrame(chains[i], columns=all_param_names)
    chain_i_df['chain'] = [i] * len(chain_i_df)
    chains_df = pd.concat([chains_df, chain_i_df], ignore_index=True)
```

```python
for name in param_names:
    print(f"Mean {name}: {np.mean(chains_df[name])}," 
          f" std dev {name}: {np.std(chains_df[name])}")
histograms = chains_df.hist(column=param_names, alpha=0.7)
for i in range(3):
    for j in range(2):
        ax = histograms[i][j]
        ax.axvline(parameters[i * 2 + j], color='r', linestyle='dashed', linewidth=1, label="True")
plt.tight_layout()
plt.legend()
plt.savefig("inference_outputs/posteriors_osc.png")
```
